# Evaluation and Benchmarking
Demonstrates confusion matrix and performance metrics.

```python

import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

np.random.seed(0)
y_true = np.random.randint(0, 3, 5000)
y_pred = y_true.copy()

flip_idx = np.random.choice(len(y_true), size=int(0.09*len(y_true)), replace=False)
y_pred[flip_idx] = (y_pred[flip_idx] + 1) % 3

cm = confusion_matrix(y_true, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.title("Confusion Matrix of RF Classifier")
plt.show()
```
