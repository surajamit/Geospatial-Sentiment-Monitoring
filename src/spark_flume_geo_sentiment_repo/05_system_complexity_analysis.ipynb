# System Complexity Analysis


```python

# System Complexity Analysis
import matplotlib.pyplot as plt
import numpy as np
# Throughput simulationtime_axis = np.arange(1, 11)throughput = np.full_like(time_axis, 1500)

plt.figure()plt.plot(time_axis, throughput)plt.title("Flume Ingestion Throughput")
plt.xlabel("Time Window")
plt.ylabel("Tweets/sec")
plt.show()
# Micro-batch latency
latency = np.full_like(time_axis, 3.2)

plt.figure()plt.plot(time_axis, latency)
plt.title("Spark Micro-batch Latency")
plt.xlabel("Batch")
plt.ylabel("Seconds")
plt.show()
```
