# Feature Engineering and KDE


```python
# Feature Engineering and KDE Heatmapimport numpy as npimport matplotlib.pyplot as pltfrom scipy.stats import gaussian_kde# synthetic geo pointsnp.random.seed(42)x = np.random.normal(77, 0.5, 200)y = np.random.normal(13, 0.5, 200)kde = gaussian_kde(np.vstack([x, y]))density = kde(np.vstack([x, y]))plt.figure()plt.scatter(x, y, c=density)plt.title("KDE Spatial Density")plt.show()
```
